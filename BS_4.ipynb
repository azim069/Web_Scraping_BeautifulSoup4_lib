```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = 'https://www.ryanscomputers.com/category/laptop-all-laptop?limit=100&osp=0'
```

```python
headers = {
  'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/109.0.0.0 Safari/537.36'
}
response = requests.get(url, headers=headers)
```

```python
soup = BeautifulSoup(response.text)
```

```python
products = soup.find_all('div',{'class':'card-body text-center'})
for product in products:
    print(product.find('p',{'class':"card-text p-0 m-0 list-view-text"}).get_text())
    print(product.find('p',{'class':"card-text p-0 m-0 list-view-text"}).find('a').get('href'))
    print(product.find('ul').find_all('li')[0].get_text())
    print(product.find('ul').find_all('li')[1].get_text())
    print(product.find('ul').find_all('li')[2].get_text())
    print(product.find('ul').find_all('li')[3].get_text())
    print(product.find('ul').find_all('li')[4].get_text())
    print(product.find('ul').find_all('li')[5].get_text())
    print(product.find('p',{"class":"pr-text cat-sp-text pb-1"}).get_text())
```
